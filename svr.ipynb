```python
import pandas as pd
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
mammals=pd.read_csv(r"C:\Users\archa\OneDrive\Desktop\Downloads\mammals.csv").sort_values('body')
mammals
```

```python
body=mammals[['body']].values
brain=mammals[['brain']].values
body
```

```python
brain
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import svm
print(body.shape)
```

```python
svm_linear=svm.SVR(kernel='linear',C=1e1)
svm_rbf=svm.SVR(kernel='rbf',C=1e1)
```

```python
svm_linear.fit(np.log(body),np.log(brain))
svm_rbf.fit(np.log(body),np.log(brain))
logfit=LinearRegression().fit(np.log(body),np.log(brain))
```

```python
mammals['lnr']=np.exp(logfit.predict(np.log(body)))
mammals['rbf']=np.exp(svm_rbf.predict(np.log(body)))
mammals['linear']=np.exp(svm_linear.predict(np.log(body)))
```
